```python



class SOM:
    def __init__(self, k, map_size=(10, 10), tau_sigma=50, tau_eta=500):
        self.k = k
        self.map = np.random.random(map_size + (self.k,))
        self.sigma0 = (map_size[0] + map_size[1]) / 4
        self.eta0 = 1
        self.tau_sigma = tau_sigma
        self.tau_eta = tau_eta

    def fit(self, X_train, epoch=10):
        X = X_train.copy()
        n = 0
        for e in range(epoch):
            np.random.shuffle(X)
            for i in range(X.shape[0]):
                d = np.sum((self.map - X[i]) ** 2, axis=2)
                winner_x, winner_y = np.unravel_index(d.argmin(), self.map.shape[:2])
                error = X[i] - self.map[winner_x, winner_y]

                sigma = self.sigma0 * np.exp(-n / self.tau_sigma)
                eta = self.eta0 * np.exp(-n / self.tau_eta)
                h = self.get_neighborhood(winner_x, winner_y, sigma)
                self.map = self.map + eta * h * (error)

                n += 1

    def get_neighborhood(self, winner_x, winner_y, sigma):
        x = np.arange(0, self.map.shape[0]) - winner_x
        y = np.arange(0, self.map.shape[1]) - winner_y
        msh = np.stack(np.meshgrid(x, y), axis=2)
        d = np.sum(msh ** 2, axis=2)
        h = np.exp(-d / (2 * (sigma ** 2)))
        h = np.tile(h, [self.map.shape[2], 1, 1]).transpose([2, 1, 0])

        return h

    def get_weights(self):
        return self.map.reshape(-1, self.map.shape[2])

```
